## 1. Setup and Data Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

fdi = pd.read_csv("../data/japan_fdi_into_india.csv")
trade = pd.read_csv("../data/india_japan_trade.csv")

fdi
```

```python
trade
```

## 2. Japanese FDI into India Over Time

First, let's look at how FDI inflows have moved year to year. I've kept the FY2025-26 figure separate since it's only a partial year (April-December) at the time of writing, not a full year like the others.

```python
fig, ax = plt.subplots(figsize=(9, 5))
colors = ["#4C72B0" if p == "full_year" else "#AAB8C2" for p in fdi["period_type"]]
bars = ax.bar(fdi["financial_year"], fdi["fdi_usd_billion"], color=colors)
ax.set_title("Japanese FDI Inflows into India", fontsize=13, weight="bold")
ax.set_ylabel("FDI (USD Billion)")
ax.set_xlabel("Financial Year")
for bar, val in zip(bars, fdi["fdi_usd_billion"]):
    ax.text(bar.get_x() + bar.get_width() / 2, val + 0.05, f"${val}B", ha="center", fontsize=9)
plt.tight_layout()
plt.show()
```

**Observation:** FDI dipped in FY2021-22 (likely still recovering from COVID-era disruption), jumped sharply in FY2023-24, then fell again in FY2024-25. This isn't a smooth growth story — it's lumpy, which makes sense once you consider that a handful of large individual investment decisions by big Japanese firms can swing the yearly total significantly.

## 3. Is FDI Growth Correlated with Trade Growth?

A natural question: when Japan invests more in India, does bilateral trade also tend to be higher in that period? To check this, I merged the two datasets on financial year and calculated the correlation between FDI inflow and total trade volume for the years where both figures are available.

```python
merged = pd.merge(
    fdi[["financial_year", "fdi_usd_billion"]],
    trade[["financial_year", "total_trade_usd_bn"]],
    on="financial_year",
    how="inner"
).dropna()

merged
```

```python
correlation = merged["fdi_usd_billion"].corr(merged["total_trade_usd_bn"])
print(f"Correlation between FDI inflow and total trade volume: {correlation:.2f}")
```

**Honest caveat:** this correlation is based on only a handful of overlapping years (the two datasets don't line up perfectly in coverage), so it's a directional signal at best, not a statistically robust result. I'm including it because it's the right way to ask the question, not because the sample size supports a strong conclusion. A proper version of this analysis would need a much longer, cleanly overlapping time series — something worth revisiting if I extend this project later.

## 4. Trade Balance: Exports vs Imports

India runs a consistent trade deficit with Japan. Let's visualize how big that gap actually is.

```python
trade_plot = trade.dropna(subset=["india_exports_to_japan_usd_bn", "india_imports_from_japan_usd_bn"])

fig, ax = plt.subplots(figsize=(9, 5))
x = trade_plot["financial_year"]
ax.bar(x, trade_plot["india_exports_to_japan_usd_bn"], label="India's Exports to Japan", color="#55A868")
ax.bar(x, -trade_plot["india_imports_from_japan_usd_bn"], label="India's Imports from Japan", color="#C44E52")
ax.axhline(0, color="black", linewidth=0.8)
ax.set_title("India-Japan Trade Balance", fontsize=13, weight="bold")
ax.set_ylabel("USD Billion")
ax.legend()
plt.tight_layout()
plt.show()
```

**Observation:** India imports roughly 3x what it exports to Japan in most years. Looking at what's actually being traded (from the source data): India's imports are dominated by electrical machinery, industrial chemicals, and precision equipment — the kind of high-value manufactured inputs India still relies on Japan for. India's exports lean toward marine products, organic chemicals, and vehicle components — lower-value-add categories by comparison. That gap says something real about where each country currently sits in the value chain, not just a trade imbalance number.

## 5. Takeaways

- Japanese FDI into India is real but inconsistent year to year — driven by lumpy, large individual investment decisions rather than steady growth.
- India's trade deficit with Japan has widened even as total trade volume has grown — both things are true at once, and the second doesn't cancel out the first.
- The kind of goods being traded (India importing higher-value manufactured goods, exporting lower-value-add products) reflects a structural gap in the relationship that raw trade totals alone don't show.

This connects directly back to what I found in my original India-Japan case study project: Japanese firms operating in India (like Toyota Kirloskar) tend to bring in capital, technology, and management practices, while India's side of the relationship is still catching up in terms of what it exports back.
